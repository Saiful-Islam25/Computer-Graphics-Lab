```python
import matplotlib.pyplot as plt

def plot_circle_points(xc, yc, x, y, points):
    # 8-way symmetry
    points.extend([
        (xc + x, yc + y),
        (xc - x, yc + y),
        (xc + x, yc - y),
        (xc - x, yc - y),
        (xc + y, yc + x),
        (xc - y, yc + x),
        (xc + y, yc - x),
        (xc - y, yc - x),
    ])

def midpoint_circle(xc, yc, r):
    x = 0
    y = r
    p = 1 - r
    points = []

    plot_circle_points(xc, yc, x, y, points)

    while x < y:
        x += 1
        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1
        plot_circle_points(xc, yc, x, y, points)

    return points

def draw_circle(xc, yc, r):
    points = midpoint_circle(xc, yc, r)
    xs, ys = zip(*points)

    plt.figure(figsize=(6, 6))
    plt.plot(xs, ys, 'bo')  # blue points
    plt.title("Midpoint Circle Drawing Algorithm")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.savefig("Midpoint Circle Drawing Algorithm.jpg")
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Example usage
center_x = 20
center_y = 20
radius = 10

draw_circle(center_x, center_y, radius)
```
