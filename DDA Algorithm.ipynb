```python
import matplotlib.pyplot as plt

def dda(x1, y1, x2, y2):
    # Calculate the difference between points
    dx = x2 - x1
    dy = y2 - y1

    # Calculate steps required for generating pixels
    steps = int(max(abs(dx), abs(dy)))

    # Calculate the increment in x & y for each step
    x_inc = dx / steps
    y_inc = dy / steps

    # Initialize starting point
    x = x1
    y = y1

    # Store the points to plot later
    points = []

    for i in range(steps + 1):
        points.append((round(x), round(y)))
        x += x_inc
        y += y_inc

    return points

def plot_line(points):
    # Unzip the list of points
    x_coords, y_coords = zip(*points)
    plt.plot(x_coords, y_coords, marker='o')
    plt.title('Line drawn using DDA Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.savefig("DDA.jpg")
    plt.grid(True)
    plt.show()
    

if __name__ == "__main__":
    x1 = int(input("Enter x1: "))
    y1 = int(input("Enter y1: "))
    x2 = int(input("Enter x2: "))
    y2 = int(input("Enter y2: "))

    points = dda(x1, y1, x2, y2)
    print("Points on the line:")
    for point in points:
        print(point)
    
    plot_line(points)
```
