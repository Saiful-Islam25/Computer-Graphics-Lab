```python
import matplotlib.pyplot as plt

def plot_circle_points(xc, yc, x, y, points):
    # 8-way symmetry
    points.extend([
        (xc + x, yc + y),
        (xc - x, yc + y),
        (xc + x, yc - y),
        (xc - x, yc - y),
        (xc + y, yc + x),
        (xc - y, yc + x),
        (xc + y, yc - x),
        (xc - y, yc - x),
    ])

def bresenham_circle(xc, yc, r):
    x = 0
    y = r
    d = 3 - 2 * r
    points = []

    while x <= y:
        plot_circle_points(xc, yc, x, y, points)
        if d < 0:
            d += 4 * x + 6
        else:
            d += 4 * (x - y) + 10
            y -= 1
        x += 1

    # Remove duplicates (optional) and return unique points
    return list(set(points))

def draw_circle(xc, yc, r):
    points = bresenham_circle(xc, yc, r)
    xs, ys = zip(*points)

    plt.figure(figsize=(6, 6))
    plt.plot(xs, ys, 'ko',color='brown')  # 'ko' means black circle markers
    plt.title("Bresenham's Circle Drawing Algorithm")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.savefig("Bresenham's Circle Drawing Algorithm.jpg")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.show()

# Example usage
center_x = 10
center_y = 10
radius = 7

draw_circle(center_x, center_y, radius)
```
