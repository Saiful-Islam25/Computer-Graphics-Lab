```python
import matplotlib.pyplot as plt
import math

def rotate_point(x, y, cx, cy, angle_degrees):
    """Rotate point (x, y) around pivot (cx, cy) by angle (degrees)."""
    angle_rad = math.radians(angle_degrees)

    # Translate point back to origin
    x -= cx
    y -= cy

    # Rotate
    x_rot = x * math.cos(angle_rad) - y * math.sin(angle_rad)
    y_rot = x * math.sin(angle_rad) + y * math.cos(angle_rad)

    # Translate back
    x_rot += cx
    y_rot += cy

    return (x_rot, y_rot)

def rotate_triangle(triangle, pivot, angle_degrees):
    return [rotate_point(x, y, pivot[0], pivot[1], angle_degrees) for x, y in triangle]

def draw_triangle(original, rotated, pivot):
    fig, ax = plt.subplots()
    plt.title("General Point-to-Point Triangle Rotation")

    # Close the triangles
    original_closed = original + [original[0]]
    rotated_closed = rotated + [rotated[0]]

    # Draw original triangle
    x_orig, y_orig = zip(*original_closed)
    ax.plot(x_orig, y_orig, 'b--', label="Original Triangle")

    # Draw rotated triangle
    x_rot, y_rot = zip(*rotated_closed)
    ax.plot(x_rot, y_rot, 'g-', linewidth=2, label="Rotated Triangle")

    # Draw pivot point
    ax.plot(pivot[0], pivot[1], 'ro', label="Pivot Point")

    ax.set_aspect('equal', adjustable='box')
    ax.grid(True)
    ax.legend()
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.savefig("Point-to-Point Rotation of a Triangle.jpg")
    plt.show()

# Example usage
triangle = [(2, 1), (4, 1), (3, 4)]
pivot = (3, 2)        # Rotation center (can be any point)
angle_degrees = 45    # Rotation angle

rotated_triangle = rotate_triangle(triangle, pivot, angle_degrees)
draw_triangle(triangle, rotated_triangle, pivot)
```
