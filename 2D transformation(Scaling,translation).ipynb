```python
import matplotlib.pyplot as plt

def translate(points, tx, ty):
    """Translate points by (tx, ty)."""
    translated_points = []
    for x, y in points:
        translated_points.append((x + tx, y + ty))
    return translated_points

def scale(points, sx, sy, pivot=(0, 0)):
    """Scale points by (sx, sy) relative to a pivot point."""
    scaled_points = []
    px, py = pivot

    for x, y in points:
        x_scaled = px + (x - px) * sx
        y_scaled = py + (y - py) * sy
        scaled_points.append((x_scaled, y_scaled))

    return scaled_points

def plot_shape(points, title):
    x_coords, y_coords = zip(*points)
    plt.plot(x_coords + (x_coords[0],), y_coords + (y_coords[0],), marker='o')  # Close the shape
    plt.title(title)
    plt.grid(True)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

if __name__ == "__main__":
    # Input shape
    points = []
    n = int(input("Enter number of points in the shape: "))
    for i in range(n):
        x = float(input(f"Enter x coordinate of point {i+1}: "))
        y = float(input(f"Enter y coordinate of point {i+1}: "))
        points.append((x, y))

    # Plot original shape
    plt.figure(figsize=(12,6))
    plt.subplot(1, 2, 1)
    plot_shape(points, "Original Shape")

    # Choose transformation
    choice = int(input("Choose transformation:\n1. Translation\n2. Scaling\nEnter choice: "))

    if choice == 1:
        tx = float(input("Enter translation along x-axis: "))
        ty = float(input("Enter translation along y-axis: "))
        new_points = translate(points, tx, ty)
    elif choice == 2:
        sx = float(input("Enter scaling factor along x-axis: "))
        sy = float(input("Enter scaling factor along y-axis: "))
        pivot_choice = input("Scale about origin (O) or a custom point (C)? [O/C]: ").strip().upper()
        if pivot_choice == 'C':
            px = float(input("Enter pivot x: "))
            py = float(input("Enter pivot y: "))
            pivot = (px, py)
        else:
            pivot = (0, 0)
        new_points = scale(points, sx, sy, pivot)
    else:
        print("Invalid choice!")
        exit()

    # Plot transformed shape
    plt.subplot(1, 2, 2)
    plot_shape(new_points, "Transformed Shape")
    plt.tight_layout()
    plt.savefig("2D transformation(Scaling,translation).jpg")
    plt.show()

    print("\nTransformed points:")
    for pt in new_points:
        print(f"({pt[0]:.2f}, {pt[1]:.2f})")
```
